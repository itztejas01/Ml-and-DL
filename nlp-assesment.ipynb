```python
import spacy
from spacy.matcher import Matcher, PhraseMatcher
nlp = spacy.load('en_core_web_sm')
matcher = Matcher(nlp.vocab)
phrase_matcher = PhraseMatcher(nlp.vocab)
```

```python
with open('owlcreek.txt') as f:
    doc1 = nlp(f.read())
```

```python
doc_sen = [sent for sent in doc1.sents]

len(doc_sen)
```

```python
print(doc_sen[1].text)
```

```python
for token in doc_sen[1]:
    print(f"{token.text:{10}} {token.pos_:{5}} {token.dep_:{10}} {token.lemma_:{10}}")
```

```python
pattern = [{'LOWER':'swimming'},{'IS_SPACE':True,'OP':"*"},{'LOWER':"vigorously"}]
```

```python
matcher.add("Swimming",[pattern])
```

```python
found_matches = matcher(doc1)
```

```python
found_matches
```

```python
def surrounding_text(doc,start,end):
    print(doc[start-10:end+5],'\n')
    
```

```python
for match_id,start,end in found_matches:
    surrounding_text(doc1,start,end)
```

```python
for sentences in doc_sen:
    for i in range(len(found_matches)):
        print(i)
        if(found_matches[i][1] < sentences.end):
            print(sentences)
            break
   
```
